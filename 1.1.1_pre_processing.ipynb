```python

import os
# Original input directory containing the images to be processed
PATH_TRAIN = '/content/drive/MyDrive/sleep monitoring project/Left Log_Thermal_Light-on_with Blanket'

# New output directory to store the preprocessed and de-duplicated images
# OUTPUT_PATH_TRAIN = '/content/drive/MyDrive/sleep monitoring project/Prone Left'

train_images = os.listdir(f'{PATH_TRAIN}')

import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim

def preprocess_image(image):
    # Convert image to grayscale (if needed)
    if len(image.shape) == 3 and image.shape[2] == 3:  # If the image is in color (3 channels)
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        # Image is already in grayscale (1 channel)
        gray_image = image

    # Resize the image to a fixed size (e.g., 224x224)
    resized_image = cv2.resize(gray_image, (224, 224), interpolation=cv2.INTER_AREA)

    return resized_image


def calculate_ssim(image1, image2):
    # Calculate Structural Similarity Index (SSIM)
    return ssim(image1, image2)

# Assuming you have a list of extracted images stored in "extracted_images_list"
# You can loop through the list and perform preprocessing and similarity comparison

# Initialize a list to store unique frames
unique_frames = []

# Initialize a similarity threshold
similarity_threshold = 0.95

# Loop through the extracted images
# for images in train_images:
count = 0
# print(f"Folder Name: {images}")
for filename in os.listdir(f"{PATH_TRAIN}/"):
    src_path = os.path.join(f"{PATH_TRAIN}/", filename)
    # Preprocess the image
    image = cv2.imread(src_path)
    processed_image = preprocess_image(image)

    # Compare the image with existing unique frames
    is_duplicate = False

    for unique_frame in unique_frames:
        similarity = calculate_ssim(processed_image, unique_frame)
        if similarity > similarity_threshold:
            is_duplicate = True
            os.remove(src_path)
            count +=1
            break

    # Add the image to the list if it's not a duplicate
    if not is_duplicate:
        unique_frames.append(processed_image)
        # Save the preprocessed image to the path with '_preprocessed' added to the filename
        output_path = src_path.replace(".jpg", ".jpg")  # Modify the extension if needed
        cv2.imwrite(output_path, processed_image)

print(f"Images has been remove duplicate: {count}")
# Now unique_frames contains the preprocessed and de-duplicated images.
# You can proceed with classification using these images.





```
